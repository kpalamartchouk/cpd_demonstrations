# Two harmonics beating

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = (16.0, 8.0)

from IPython.html.widgets import interact
from IPython.display import display, Audio
```

```python
@interact(f1=(435.0,445.0), a1=(0.5, 1.5))
def beat_freq(f1=441.0, a1=1.0, plot=True):
    f0 = 440.0
    a0 = 1.0
    max_time = 15
    max_time_to_plot = 2
    max_time_to_plot_close = 0.02
    rate = 8000
    times = np.linspace(0, max_time,rate*max_time)
    signal0 = a0*np.sin(2*np.pi*f0*times)
    signal1 = a1*np.sin(2*np.pi*f1*times)
    signal = signal0 + signal1

    print(f0, f1, abs(f0-f1))
    display(Audio(data=signal, rate=rate))
    if plot:
        plt.plot(signal0[0:max_time_to_plot_close*rate])
        plt.plot(signal1[0:max_time_to_plot_close*rate])
        plt.show()
        
        plt.plot(signal[0:max_time_to_plot*rate])
        plt.show()
    return signal
```
